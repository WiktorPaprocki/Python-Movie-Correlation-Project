```python
# Creating correct year column

df[['released_date', '1']] = df['released'].str.split('(', expand = True)

df = df.drop(columns = '1')
df['released_date'] = pd.to_datetime(df['released_date'])
df['released_year'] = df['released_date'].dt.year
df['released_year'] = df['released_year'].astype('string')
df['released_year'] = df['released_year'].astype('string').str[:4]
df
```

```python
df.drop_duplicates()
```

```python
# Budget vs Gross Correlation

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')

plt.show()
```

```python
# Budget vs Gross plot using seaborn\
sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {'color': 'red'}).set(title='Budget vs Gross Earnings')

```
